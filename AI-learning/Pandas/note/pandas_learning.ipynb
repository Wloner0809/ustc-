#### Python基础(一些语法糖)

```python
# 列表推导式(支持多层嵌套)
print([m + ' ' + n for m in ['ustc', 'nico'] for n in ['edu', 'com']]) 
# 带if选择的条件赋值
value = '666' if 2 > 1 else '444'
print(value)

# 匿名函数
print([(lambda x: x * x)(i) for i in range(10)])
# 用map函数实现, 需要调用list()函数转换为列表
print(list(map(lambda x: x * x, range(10))))

# zip函数将可迭代对象打包成元组, 通常用于循环
L1 = [1, 2, 3, 4]
L2 = ['a', 'b', 'c', 'd']
print(list(zip(L1, L2)))
# 解压操作zip(*)
print(list(zip(*zip(L1, L2))))
for (x, y) in zip(L1, L2):
    print(x, y)
# enumerate函数在迭代时绑定遍历序号, 通常用于循环
print('enumerate result:')
for index, value in enumerate(L2):
    print(index, value)
# 对两个列表建立字典映射时, 可以用zip函数
print(dict(zip(L1, L2)))
```

#### Numpy基础

```python
# numpy数组构造
import numpy as np
print(np.array([1, 2, 3]))
# 生成等差数列
print(np.arange(0, 10, 2)) # 起始值, 终止值, 步长
print(np.linspace(1, 5, 11)) # 起始值, 终止值, 样本个数
# 特殊矩阵
print('全0矩阵')
print(np.zeros((2, 3))) # 全0矩阵
print('单位矩阵')
print(np.eye(3)) # 3*3单位矩阵
print(np.eye(3, k=1)) # 3*3单位矩阵, 偏移主对角线为1
print('对角矩阵')
print(np.diag([1, 2, 3])) # 对角矩阵
print('全10矩阵')
print(np.full((2, 3), 10)) # 全10矩阵
print('随机矩阵')
print(np.random.rand(2, 3)) # 随机矩阵0-1均匀分布
print(np.random.randn(2, 3)) # 随机矩阵标准正态分布
print(np.random.randint(0, 10, (2, 3))) # 随机矩阵整数均匀分布(2, 3)为维度
print(np.random.choice([10, 20, 30], (3, 3), p=[0.4, 0.2, 0.4]))# 以一定概率进行选择
print(np.random.permutation(['ustc', 'nico', 'edu', 'com'])) # 打散原列表
```

```python
# numpy数组变形、合并等操作
# 转置
print(np.zeros((2,4)).T)
# 合并
print(np.r_[np.zeros((2, 3)), np.ones((2, 3))]) # 按行合并
print(np.c_[np.zeros((2, 3)), np.ones((2, 3))]) # 按列合并
print(np.c_[np.array([1,1]), np.zeros((2, 3))]) # 一维二维数组合并, 按列合并
# 维度变换 
print(np.random.randn(6).reshape((3, 2))) # 2*3变为3*2
print(np.random.randn(2,3).reshape(-1)) # reshape(-1)将多维数组变为一维数组, 允许一个维数空缺
# 数组切片
print(np.arange(9).reshape(3, 3))
print(np.arange(9).reshape(3, 3)[1:3, 1:3]) # 二维数组切片
print(np.arange(9).reshape(3, 3)[1:3, :]) # 二维数组切片
print(np.arange(9).reshape(3, 3)[np.ix_([1,2], [True, False, True])]) # 在对应的维度上使用布尔索引, 此时不能使用slice切片
```

```python
# 常用函数
# where函数指定满足条件和不满足条件的填充值
print(np.where(np.arange(9).reshape(3, 3) > 0, 1, 0))
# 返回索引函数: nonzero, argmax, argmin
# nonzero返回非零元素的索引
print(np.nonzero(np.arange(9)))
# argmax返回最大值的索引
print(np.argmax(np.arange(9)))
# argmin返回最小值的索引
print(np.argmin(np.arange(9)))
# any函数: 当序列至少有一个True或非零元素时返回True
print(np.any(np.arange(2)))
# all函数: 当序列所有元素都为True或非零元素时返回True
print(np.all(np.arange(2)))
print(np.arange(1, 5))
# cumprod函数: 返回给定序列的累积乘法结果
print(np.cumprod(np.arange(1, 5)))
# cumsum函数: 返回给定序列的累积加法结果
print(np.cumsum(np.arange(1, 5)))
# diff函数: 返回给定序列的累积差值, 返回长度比原序列少1
print(np.diff(np.arange(1, 5)))
# 统计函数: sum, mean, std, var, max, min, median, average, quantile(全局函数, 不能通过array.quantile()调用)
print(np.quantile(np.arange(1, 5), 0.5)) # 返回给定序列的分位数, 这里是中位数
# 忽略有缺失值数组
print(np.array([1, 2, np.nan, 3]))
print(np.nansum(np.array([1, 2, np.nan, 3])))
print('协方差、相关系数的计算')
print(np.cov(np.array([1, 3, 5, 9]), np.array([1, 5, 3, -9]))) # 协方差
print(np.corrcoef(np.array([1, 2, 5]), np.array([2, 5, 6]))) # 相关系数
# 二维NUmpy数组的axis参数: axis=0表示列的统计指标; axis=1表示行的统计指标
print(np.arange(9).reshape(3, 3).sum(axis=0))
print(np.arange(9).reshape(3, 3).sum(axis=1))
```

```python
# 广播机制处理不同维度数组之间的操作
# 标量和数组的操作, 标量自动把大小扩充为数组大小
print(3 * np.random.rand(2, 3) + 2)
# 二维数组之间的操作, 其中一个维度实1*m或者n*1, 则自动扩充为m*n
print(np.ones((2, 3)) + np.zeros((1, 3)))
# 一维数组与二维数组的操作, 将一维数组视作1*m的二维数组
print(np.arange(3) + np.ones((3, 3)))
# 向量内积的计算
print(np.arange(3).dot(np.arange(2, 5)))
# 向量范数和矩阵范数
print(np.linalg.norm(np.arange(4).reshape(2, 2), 'fro'))
# 矩阵乘法
print(np.arange(4).reshape(2, 2) @ (np.arange(4).reshape(2, 2)))
```

```python
import numpy as np
import pandas as pd
df = pd.read_csv('../data/learn_pandas.csv')
# 提取前7列
df = df[df.columns[:7]]
# 替换函数
print('替换函数')
print(df['Gender'].replace({'Female':0, 'Male':1}).head()) # 通过字典替换
print(df['Gender'].replace(['Female', 'Male'], [0, 1]).head()) # 通过列表替换
series = pd.Series([1, 2, 'a', 1, 2, 'b'])
print(series.replace([1, 2], method='bfill')) # method参数表示用后一个最近未被替换的值进行替换
print(series.replace([1, 2], method='ffill')) # method参数表示用前一个最近未被替换的值进行替换
# 逻辑替换
print('逻辑替换')
series = pd.Series([-1, 1.2345, 100, -50])
print(series.where(series<0)) # where函数返回满足条件的值, 不指定值时不满足的值用NaN填充
print(series.where(series<0, 0)) # where函数返回满足条件的值, 指定值时不满足的值用指定值填充
print(series.mask(series<0)) # mask函数返回不满足条件的值, 不指定值时满足的值用NaN填充
print(series.mask(series<0, 0)) # mask函数返回不满足条件的值, 指定值时满足的值用指定值填充
# 数值替换
print('数值替换')
print(series)
print(series.round(2)) # round函数返回指定精度的四舍五入值
print(series.abs()) # abs函数返回绝对值
print(series.clip(0, 2)) # clip函数返回截断值, 小于0的值替换为0, 大于2的值替换为2
# 排序函数
print('排序函数')
df_test = df[['Grade', 'Name', 'Height', 'Weight']].set_index(['Grade', 'Name']) # 把年级和名字当成索引
print(df_test.sort_values('Height').head()) # sort_values函数返回按指定列排序的结果
print(df_test.sort_values('Height', ascending=False).head()) # sort_values函数返回按指定列排序的结果, ascending参数表示升序或降序
# 多列排序
print('多列排序')
print(df_test.sort_values(['Weight', 'Height'], ascending=[True, False]).head()) # sort_values函数返回按指定列排序的结果, ascending参数表示升序或降序
# 索引排序
print('索引排序')
print(df_test.sort_index(level=['Grade', 'Name'], ascending=[True, False]).head()) # sort_index函数返回按指定索引排序的结果, level参数表示索引层级, ascending参数表示升序或降序
# apply方法, 用于DataFrame的行/列的迭代
print('apply方法')
df_demo = df[['Height', 'Weight']]
print(df_demo.apply(lambda x: x.mean())) # apply函数返回对每一列应用函数的结果
print(df_demo.apply(lambda x: x.mean(), axis=1).head()) # apply函数返回对每一行应用函数的结果
```

#### 窗口对象

```python
import numpy as np
import pandas as pd
# 滑窗对象rolling
print('滑窗对象')
series = pd.Series([1, 2, 6, 16, 30])
roller = series.rolling(window=2) # window参数表示窗口大小
print(roller)
print('聚合函数的使用')
print(roller.mean()) # mean函数返回滑窗平均值
print('均值使用apply实现')
print(roller.apply(lambda x: x.mean())) # apply函数返回滑窗平均值
print(roller.sum()) # sum函数返回滑窗和
print(roller.cov(series)) # cov函数返回滑窗协方差
print(roller.corr(series)) # corr函数返回滑窗相关系数
# 下面三个类滑窗函数的参数n可以为负值, 代表反方向操作
print(series.shift(3)) # 取向前第n个元素
print(series.diff(1)) # 与向前第n个元素的差值
print(series.pct_change(3)) # 与向前第n个元素相比计算增长率

# 扩张窗口
print('扩张窗口')
print(series.expanding(3).sum()) # expanding函数返回扩张窗口的和(窗口逐步扩张)
```

```python
import numpy as np
import pandas as pd
# 列索引, 一般通过[]实现
print('列索引')
df = pd.read_csv('../data/learn_pandas.csv', usecols=['School', 'Grade', 'Name', 'Gender', 'Weight', 'Transfer'])
print(df['Name'].head())
print(df.Name.head()) # 通过.实现(要求列名不能有空格)
print(df[['School', 'Name']].head()) # 取出多个

# 行索引
print('行索引')
series = pd.Series([1, 2, 3, 4, 5], index=['a', 'b', 'c', 'a', 'e']) # 以字符串为索引的series
print(series['a']) # 通过[item]实现单个索引对应元素
print(series[['a', 'b']]) # 通过[item1, item2]实现多个索引对应元素
print(series['b':'e']) # 通过[item1:item2]实现切片索引对应元素,要求索引唯一出现
print(series.sort_index()['a':'b']) # 有重复元素排序后可以使用切片索引
series = pd.Series(['a', 'b', 'c', 'd', 'e', 'f'], index=[1, 3, 1, 2, 5, 4]) # 以数字为索引的series(默认就是以数字为索引)
print(series[1]) # 通过[item]实现单个索引对应元素
print(series[0:5:2]) # 通过[item1:item2:step]实现切片索引对应元素

# loc索引器, 基于元素, loc[index, column], 省略column表示取出所有列进行行的筛选
print('loc索引器')
df_test = df.set_index('Name')
print(df_test.head())
# 对单个元素进行操作
print('重名')
print(df_test.loc['Qiang Sun']) # 重名
print('无重名')
print(df_test.loc['Quan Zhao']) # 无重名
print(df_test.loc['Qiang Sun', 'Gender']) # 同时选择行和列
# 对元素列表进行操作
print(df_test.loc[['Qiang Sun', 'Quan Zhao'], ['Gender', 'Weight']])
# 对元素切片进行操作
print(df_test.loc['Gaojuan You':'Gaoqiang Qian', 'School':'Gender'])
# 对布尔列表进行操作, 可以通过与或非操作得到布尔列表
print(df_test.loc[df_test.Weight>60].head())
print(df_test.loc[df_test.Grade.isin(['Freshman', 'Senior'])].head()) # 利用isin()方法构造布尔列表
# 对函数进行操作
print(df_test.loc[lambda x: x.Weight<60].head()) 
print(df_test.loc[lambda x: slice('Gaojuan You', 'Gaoqiang Qian')]) # 利用切片构造函数
```

```python
import numpy as np
import pandas as pd
# iloc索引器, 基于位置(与loc类似)
print('iloc索引器')
df = pd.read_csv('../data/learn_pandas.csv', usecols=['School', 'Grade', 'Name', 'Gender', 'Weight', 'Transfer'])
df_test = df.set_index('Name')
# 对单个元素进行操作
print(df_test.iloc[1, 2]) # 同时选择行和列
# 对元素列表进行操作
print(df_test.iloc[[1, 2], [2, 3]])
# 对元素切片进行操作
print(df_test.iloc[1:3, 2:4]) # 不包含结束端点
print(df_test.iloc[lambda x: slice(1, 4)]) 
# 对布尔列表进行操作
print(df_test.iloc[(df_test.Weight>60).values].head()) # 注意传入的是values

# query方法
print('query方法')
print(df.query('((School == "Fudan University")&'
         ' (Grade == "Senior")&'
         ' (Weight > 70))|'
         '((School == "Peking University")&'
         ' (Grade != "Senior")&'
         ' (Weight > 80))'))
print(df.query('Weight > Weight.mean()').head())
print(df.query('Grade not in ["Freshman", "Senior"]').head()) # 英语字面用法
print(df.query('Grade == ["Freshman", "Senior"]').head()) # ==表示元素出现在列表中
low, high = 60, 80
print(df.query('(Weight > @low) & (Weight < @high)').head()) # @表示引用外部变量

# 随机抽样(行可以看成一个样本, 列可以看成一个特征)
print('随机抽样')
df_sample = pd.DataFrame({'id' : list('abcde'), 'value' : [1, 2, 3, 4, 5]})
print(df_sample)
print(df_sample.sample(n=3, replace=True, weights=df_sample.value)) # n参数表示抽样的数量, replace参数表示是否放回抽样, weights参数表示抽样的概率, axis表示抽样方向, frac表示抽样比例
```

#### 聚合函数

```python
import numpy as np
import pandas as pd
# 内置聚合函数
print('内置聚合函数')
df = pd.read_csv('../data/learn_pandas.csv')
gb = df.groupby('Gender')['Height']
print(gb.idxmin()) # idxmin()方法返回最小值的索引
print(gb.quantile(0.8)) # quantile()方法返回分位数
gb = df.groupby('Gender')[['Height', 'Weight']] # 多列分组
print(gb.max()) # max()方法返回最大值, 按照列迭代计算

# agg方法
print('agg方法')
print(gb.agg(['sum', 'idxmax', 'skew'])) # 使用多个函数, agg()方法返回多个聚合结果
print(gb.agg({'Height':['mean', 'max'], 'Weight':'count'})) # 使用字典, 对特定的列使用特定的聚合函数
print(gb.agg(lambda x: x.max()-x.min())) # 使用函数, 函数的参数是数据源的列, 逐列进行计算
print(gb.agg([('range', lambda x: x.max()-x.min()), ('my_func', 'mean')])) # 使用元组对聚合结果重命名, 元组的第一个元素是列名, 第二个元素是函数
print(gb.agg({'Height': [('my_func', 'mean'), 'max'], 'Weight': [('range', lambda x: x.count())]})) # 使用字典对聚合结果重命名, 重命名需要加方括号
```

```python
import numpy as np
import pandas as pd

# 长表, 如果把某一特征存储在某一列中, 则就是关于该特征的长表
print(pd.DataFrame({'Gender': ['F', 'M', 'M', 'F'], 'Height': [163, 178, 173, 165]}))
# 宽表, 如果把某一特征存储在多列中(列中的元素是某一其他的相关特征数值), 则就是关于该特征的宽表
print(pd.DataFrame({'Height: F':[163, 160], 'Height: M':[178, 173]}))

# pivot方法, 需要满足唯一性需求(即行列索引组合唯一),否则报错
print('pivot方法')
df = pd.DataFrame({'Class': [1,2,1,2], 'Name':['San Zhang', 'San Zhang', 'Si Li', 'Si Li'], 'Subject':['Chinese', 'Math', 'Chinese', 'Math'], 'Grade':[80, 75, 90, 85]})
print(df)
print(df.pivot(index='Name', columns='Subject', values='Grade')) # pivot()方法返回以index的unique值为行索引, columns的unique值为列索引, values为值的表
df = pd.DataFrame({'Class':[1,1,2,2,1,1,2,2],'Name':['San Zhang', 'San Zhang', 'Si Li', 'Si Li', 'San Zhang', 'San Zhang', 'Si Li', 'Si Li'], 'Examination': ['Mid', 'Final', 'Mid', 'Final', 'Mid', 'Final', 'Mid', 'Final'], 'Subject':['Chinese', 'Chinese', 'Chinese', 'Chinese', 'Math', 'Math', 'Math', 'Math'], 'Grade':[70, 80, 65, 90, 85, 75, 60, 88], 'rank':[10, 15, 20, 14, 20, 7, 6, 3]})
print(df)
print(df.pivot(index=['Class', 'Name'], columns=['Subject', 'Examination'], values=['Grade', 'rank'])) # 传入列表, 生成多级索引

# pivot_table方法
print('pivot_table方法')
# 行列索引组合不唯一的情况
df = pd.DataFrame({'Name':['San Zhang', 'San Zhang', 'San Zhang', 'San Zhang', 'Si Li', 'Si Li', 'Si Li', 'Si Li'], 'Subject':['Chinese', 'Chinese', 'Math', 'Math', 'Chinese', 'Chinese', 'Math', 'Math'], 'Grade':[80, 75, 90, 85, 100, 99, 87, 91]})
print(df)
print(df.pivot_table(index='Name', columns='Subject', values='Grade', aggfunc='mean')) # pivot_table()方法返回以index的unique值为行索引, columns的unique值为列索引, values为值的表, aggfunc参数表示聚合函数(使相同的行列组合对应的多个值变为一个值)
print(df.pivot_table(index='Name', columns='Subject', values='Grade', aggfunc='mean', margins=True)) # pivot_table具有边际汇总的功能, 边际汇总的聚合方式与aggfunc相同 

# melt方法, 将宽表转为长表
print('melt方法')
df = pd.DataFrame({'Class':[2,1], 'Name':['Terence', 'Lily'], 'English':[90,91], 'Math':[89, 92]})
print(df)
print(df.melt(id_vars=['Class', 'Name'], value_vars=['English', 'Math'], var_name='Subject', value_name='Grade')) # id_vars参数表示保持不变的列, value_vars参数表示要转换的列, var_name参数表示转换后的列名, value_name参数表示转换后的值名
df_melt = df.melt(id_vars=['Class', 'Name'], value_vars=['English', 'Math'], var_name='Subject', value_name='Grade')
print(df_melt.pivot(index=['Class','Name'], columns='Subject', values='Grade')) # 逆操作
df_melt = df_melt.pivot(index=['Class','Name'], columns='Subject', values='Grade')
df_melt = df_melt.reset_index().rename_axis(columns={'Subject':''}) # 重置索引, 重命名列索引
print(df_melt)

# wide_to_long方法, 将宽表转为长表(适用于列中包含了交叉类别)
print('wide_to_long方法')
df = pd.DataFrame({'Class':[2,1], 'Name':['Terence', 'Lily'], 'English_Mid':[90,91], 'Math_Mid':[89, 92], 'English_Final':[92, 93], 'Math_Final':[94, 95]})
print(df)
print(pd.wide_to_long(df, stubnames=['English', 'Math'], i=['Class', 'Name'], j='Examination', sep='_', suffix='.+')) # stubnames参数表示列名的前缀, i参数表示保持不变的列, j参数表示转换后的列名, sep参数表示分隔符, suffix参数表示列名的后缀
```

#### 关系型连接

```python
import numpy as np
import pandas as pd
# 值连接
print('值连接')
df1 = pd.DataFrame({'Name': ['San Zhang', 'Si Li'], 'Age': [20, 21]})
df2 = pd.DataFrame({'Name': ['Wu Wang', 'Si Li'], 'Gender': ['M', 'F']})
# merge方法, on参数表示连接的列, how参数表示连接方式, left表示左连接, right表示右连接, inner表示内连接, outer表示外连接
print(df1.merge(df2, on='Name', how='left')) 
print(df1.merge(df2, on='Name', how='right'))
print(df1.merge(df2, on='Name', how='inner'))
print(df1.merge(df2, on='Name', how='outer'))
df1 = pd.DataFrame({'df1_name': ['San Zhang', 'Si Li'], 'Age': [20, 21]})
df2 = pd.DataFrame({'df2_name': ['Wu Wang', 'Si Li'], 'Gender': ['M', 'F']})
print(df1.merge(df2, left_on='df1_name', right_on='df2_name', how='left')) # 连接的列不具备相同的列名
df1 = pd.DataFrame({'Name':['Lily'], 'Grade':[80]})
df2 = pd.DataFrame({'Name':['Lily'], 'Grade':[90]})
print(df1.merge(df2, on='Name', how='left', suffixes=['_English', '_Math'])) # suffixes参数表示连接的列名相同时的后缀
df1 = pd.DataFrame({'Name':['San Zhang', 'San Zhang'], 'Age':[20, 21], 'Class':['one', 'two']})
df2 = pd.DataFrame({'Name':['San Zhang', 'San Zhang'], 'Gender':['F', 'M'], 'Class':['two', 'one']})
print(df1)
print(df2)
print('Wrong Answer')
print(df1.merge(df2, on='Name', how='left')) # 返回笛卡尔积
print('Right Answer')
print(df1.merge(df2, on=['Name', 'Class'], how='left')) # 多列连接, 返回正确结果

# 索引连接
print('索引连接')
df1 = pd.DataFrame({'Age': [20, 21]}, index=pd.Index(['San Zhang', 'Si Li'], name='Name'))
df2 = pd.DataFrame({'Gender': ['M', 'F']}, index=pd.Index(['Wu Wang', 'Si Li'], name='Name'))
print(df1.join(df2, how='left'))
df1 = pd.DataFrame({'Grade':[70]}, index=pd.Series(['San Zhang'], name='Name'))
df2 = pd.DataFrame({'Grade':[80]}, index=pd.Series(['San Zhang'], name='Name'))
print(df1.join(df2, lsuffix='_Chinese', rsuffix='_Math')) # suffixes参数表示连接的列名相同时的后缀
# 以多列为键操作, 需要多级索引
df1 = pd.DataFrame({'Age':[20, 21]}, index=pd.MultiIndex.from_arrays([['San Zhang', 'San Zhang'], ['one', 'two']], names=('Name', 'Class')))
df2 = pd.DataFrame({'Gender':['F', 'M']}, index=pd.MultiIndex.from_arrays([['San Zhang', 'San Zhang'], ['two', 'one']], names=('Name', 'Class')))
print(df1.join(df2))
```

# 缺失数据

#### 缺失值的统计和删除

```python
import numpy as np
import pandas as pd
# 缺失信息的统计
print('缺失信息的统计')
df = pd.read_csv('../data/learn_pandas.csv', usecols=['Grade', 'Name', 'Gender', 'Height', 'Weight', 'Transfer'])
print(df.isna().head()) # isna方法, 与isnull方法等价
print(df.isna().mean()) # mean方法返回每一列缺失值的比例
print(df[df.Height.isna()].head()) # 用Series上的isna方法布尔索引找到某一列的缺失, 这里找的是身高缺失的行
df_tmp = df[['Height', 'Weight', 'Transfer']]
print(df[df_tmp.isna().all(1)].head()) # isna结合all判断是否全部缺失
print(df[df_tmp.isna().any(axis=1)].head()) # isna结合any判断是否至少有一个缺失, axis=1不可以缺失否则报错TypeError: DataFrame.any() takes 1 positional argument but 2 were given
print(df[df_tmp.notna().all(1)].head()) # notna结合all判断是否没有缺失

# 缺失信息的删除
print('缺失信息的删除')
print(df.shape)
print(df.dropna(how='any', subset=['Height', 'Weight']).shape) # dropna方法, how参数表示删除的方式, subset参数表示删除的列
print(df.dropna(axis=1, thresh=df.shape[0]-15).head()) # thresh表示删除的非缺失值个数阈值, 非缺失值个数小于阈值的列会被删除
print(df.dropna(axis=1, thresh=df.shape[0]-15).shape)
```

#### 缺失值的填充和插值


```python
import numpy as np
import pandas as pd
# 利用fillna方法填充缺失值
print('利用fillna方法填充缺失值')
series = pd.Series([np.nan, 1, np.nan, np.nan, 3], index=list('abaae'))
print(series)
print(series.fillna(method='ffill')) # ffill表示用前一个非缺失值向后填充
print(series.fillna(method='ffill', limit=1)) # limit参数表示最多填充的个数, 连续的缺失值只填充一个
print(series.fillna(value=series.mean())) # value参数表示填充的值
print(series.fillna({'a': 10})) # 传入字典, 指定填充的值

# 插值函数
print('插值函数')
series = pd.Series([np.nan, np.nan, 1, np.nan, np.nan, np.nan, 2, np.nan, np.nan])
print(series.values)
# 线性插值
print(series.interpolate(limit_direction='backward', limit=1).values) # limit_direction参数表示插值的方向, limit参数表示最多插值的个数, 默认线性插值
print(series.interpolate(limit_direction='both', limit=1).values)
# 最近邻插值, 注意这里要装scipy包
print(series.interpolate(method='nearest').values)
# 索引插值
series = pd.Series([0, np.nan, 100], index=[0, 1, 10])
print(series)
print(series.interpolate()) # 默认线性插值
print(series.interpolate(method='index')) # 索引插值, 计算相应索引大小对应的值
series = pd.Series([0, np.nan, 100], index=[pd.Timestamp('20200101'), pd.Timestamp('20200102'), pd.Timestamp('20200111')])
print(series)
print(series.interpolate()) # 默认线性插值
print(series.interpolate(method='index')) # 对时间戳索引进行索引插值, 计算相应索引大小对应的值
```

#### Nullable类型

```python
import numpy as np
import pandas as pd
# 缺失记号及其缺陷
print('缺失记号及其缺陷')
print(None==None, None==[], None==False) # None只与自身相等
print(np.nan==np.nan, np.nan==[], np.nan==False) # np.nan与任何对象都不相等
s1 = pd.Series([1, np.nan])
s2 = pd.Series([1, 2])
s3 = pd.Series([1, np.nan])
print(s1.equals(s2), s1.equals(s3)) # equals方法, 比较两个Series是否相等, 这里直接跳过了np.nan的比较
print(pd.to_datetime(['20231124', np.nan])) # 时间序列用pd.NaT表示时间戳的缺失值

# Nullable类型的性质
print('Nullable类型的性质')
print(pd.Series([1, np.nan], dtype='Int64'))
print(pd.Series([True, np.nan], dtype='boolean'))
print(pd.Series([1, np.nan], dtype='Int64') + 1)
print(pd.Series([1, np.nan], dtype='Int64') == 1)
print(pd.Series([1, np.nan], dtype='Int64') + 0.5)
s = pd.Series(['a', 'b'])
print(s[pd.Series([True, np.nan]).astype('boolean')]) # astype方法, 转换类型为boolean, 这里把np.nan转换为了False
print(pd.Series([True, np.nan]).astype('boolean') & True) # boolean类型会根据逻辑运算是否能确定唯一结果来返回相应的值
df = pd.read_csv('../data/learn_pandas.csv')
df = df.convert_dtypes() # convert_dtypes方法, 将DataFrame中的列转换为Nullable类型
print(df.dtypes)

# 缺失数据的计算和分组
print('缺失数据的计算和分组')
s = pd.Series([1, np.nan, 2, 4, 3])
print(s.sum()) # 加法np.nan当作0
print(s.prod()) # 乘法np.nan当作1
print(s.cumsum()) # 累加np.nan被跳过
```
